#### **Completitud**

```python
# Lista para almacenar los resultados
resultado_covid = []

# Bucle para iterar sobre la base de datos
for columna in covid.columns:
    valores_nulos = covid[columna].isnull().sum()  # Número de valores nulos en la columna
    longitud = len(covid[columna])  # Longitud total de la columna
    porcentaje = (valores_nulos / longitud) * 100  # Porcentaje de valores nulos

    # Guardar el nombre de la columna y su porcentaje de completitud
    resultado_covid.append({'Columna': columna, 'Porcentaje de COMPLETITUD': 100 - round(porcentaje,2)})

# Crear una base de datos para mostrar una tabla
tabla_completitud = pd.DataFrame(resultado_covid)

# Imprimir la tabla 
print(tabla_completitud)
```

```python
# Lista para almacenar los resultados
resultado_poblacion = []

# Bucle para iterar sobre la base de datos
for columna in poblacion.columns:
    valores_nulos = poblacion[columna].isnull().sum()  # Número de valores nulos en la columna
    longitud = len(poblacion[columna])  # Longitud total de la columna
    porcentaje = (valores_nulos / longitud) * 100  # Porcentaje de valores nulos

    # Guardar el nombre de la columna y su porcentaje de completitud
    resultado_poblacion.append({'Columna': columna, 'Porcentaje de COMPLETITUD': 100 - round(porcentaje,2)})

# Crear una base de datos para mostrar una tabla
tabla_completitud_poblacion = pd.DataFrame(resultado_poblacion)

# Imprimir la tabla 
print(tabla_completitud_poblacion)
```

##### Variable de **COUNTRY/REGION**

```python
# Numero esperado de localizaciones
paises_poblacion = len(poblacion['Country/Region'].unique())

# Obtener la cantidad de países únicos en la comlumna 'Country/Region' con respecto de la de poblacion
paises_covid = len(covid['Country/Region'].unique())

# Verificar si la cantidad de países únicos es menor que la esperada
if paises_covid < paises_poblacion:
    resultado = (paises_covid * 100) / paises_poblacion
    print(f'Porcentaje de RAZONABILIDAD de “Country/Region": {resultado: .2f}%')

else:
    print(f'Porcentaje de RAZONABILIDAD de "Country/Region" es del 100%') 
```

#### **Identificabilidad**

```python
# Lista para almacenar los resultados
resultado_ide_covid = []

# Bucle para iterar sobre la base de datos
for columna in covid.columns:
    valores_unicos = len(covid[columna].unique()) # Número de valores unicos en la columna
    valores_totales = len(covid[columna]) # Longitud total de la columna
    porcentaje_ide = (valores_unicos/valores_totales) * 100 # Porcentaje de valores unicos

    # Guardar el nombre de la columna y su porcentaje de identificabilidad
    resultado_ide_covid.append({'Columna': columna, 'Porcentaje de IDENTIFICABILIDAD': round(porcentaje_ide,2)})

# Crear una base de datos para mostrar una tabla
tabla_identificabilidad_covid = pd.DataFrame(resultado_ide_covid)

# Imprimir la tabla
print(tabla_identificabilidad_covid)
```

```python
# Lista para almacenar los resultados
resultado_ide_poblacion = []

# Bucle para iterar sobre la base de datos
for columna in poblacion.columns:
    valores_unicos = len(poblacion[columna].unique()) # Número de valores unicos en la columna
    valores_totales = len(poblacion[columna]) # Longitud total de la columna
    porcentaje_ide = (valores_unicos/valores_totales) * 100 # Porcentaje de valores unicos
   
    # Guardar el nombre de la columna y su porcentaje de identificabilidad
    resultado_ide_poblacion.append({'Columna': columna, 'Porcentaje de IDENTIFICABILIDAD': round(porcentaje_ide,2)})

# Crear una base de datos para mostrar una tabla
tabla_identificabilidad_poblacion = pd.DataFrame(resultado_ide_poblacion)

# Imprimir la tabla
print(tabla_identificabilidad_poblacion)
```
